```python
import numpy as np
import pandas as pd
from osgeo import gdal
import matplotlib.pyplot as plt
from PIL import ImageColor
from PIL import Image
```

```python
img = gdal.Open('/content/drive/MyDrive/saocom.tif/subset_2_of_20201209Z_hh_roi _Spk.tif')
```

```python
imgread = img.ReadAsArray().astype(np.float)
[x,y] = imgread.shape
```

```python
print(x)
print(y)
```

```python
matrix =[]
for i in range(x):
  for j in range(y):
    value = imgread[i,j]
    if (value>0.1)&(value<0.25):
      
      matrix.append(0)
     
    elif (value>0)&(value<0.1):
      value1 = 100
      matrix.append(value1)
    elif (value>0.25)&(value<0.40):
      matrix.append(150)
    elif (value>0.4)&(value<0.6):
      matrix.append(200)  
    else:
      matrix.append(255)  

```

```python
m = Image.new('RGBA',(742,1231))
K= 0
for i in range(x):
  for j in range(y):
    value = imgread[i,j]
    if (value >= 0.1)&(value<=0.2) :
      
      
      m.putpixel((i,j),ImageColor.getcolor('red','RGBA'))
    elif (value>=0) &(value<0.1):
      m.putpixel((i,j),ImageColor.getcolor('gray','RGBA'))
    elif (value>0.2) & (value<=0.3)  :
      m.putpixel((i,j),ImageColor.getcolor('blue','RGBA'))
    elif (value>0.3) & (value<=0.42)  :
      m.putpixel((i,j),ImageColor.getcolor('violet','RGBA'))



    else :
      m.putpixel((i,j),ImageColor.getcolor('green','RGBA'))
K =+1
```
