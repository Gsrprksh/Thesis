```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from osgeo import gdal
```

```python
x = gdal.Open('/content/drive/MyDrive/final folder/landsat/subset verano_utm.tif')
x_array = x.ReadAsArray()
x_array.shape
x_transforms = x.GetGeoTransform()
x_projection = x.GetProjection()
```

```python
a,b = x_array.shape
b
```

```python
plt.imshow(x_array==22)
```

```python
x_1 = x_array[1:,:1231]
p,q =x_1.shape
p
```

```python
new_pixel = np.zeros([p,q])
for i in range(p):
  for j in range(q):
    pixel =x_1[i,j]
    if pixel ==11:
      new_pixel[i,j] = 250
    else:
      new_pixel[i,j] =0
plt.imshow(new_pixel)
```

```python
drive = gdal.GetDriverByName('GTiff')
drive.Register()
drive1 = drive.Create('verano11.tif',xsize = q,ysize = p ,bands=1 )
#drive2 = drive1.GetRasterBand(1)
drive1.SetGeoTransform(x_transforms)
drive1.SetProjection(x_projection)
drive2 = drive1.GetRasterBand(1)
drive2.WriteArray(new_pixel)
drive2.SetNoDataValue(np.nan)
drive2.FlushCache()
drive2 = None
drive1 =None
```
